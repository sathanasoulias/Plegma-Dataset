### Electric data Plots


```python
house = 'House_01'
month = '2023-08'


# Parse data
# ----------
electric_data = pd.read_csv('clean_dataset/'+house+'/Electric_data/'+month+'.csv',header=0)
electric_data['timestamp'] = pd.to_datetime(electric_data['timestamp'])
electric_data.set_index('timestamp', inplace=True)

# Discard the V,A,issues columns
# ------------------------------
columns_to_discard = ['V','A','issues']  
df                 = electric_data.drop(columns=[col for col in columns_to_discard if col in electric_data.columns])


def make_plot(df):
    """
    Creates a line plot for time series data using Plotly.

    This function takes a DataFrame as input and generates a line plot for each column, excluding specific columns ('V', 'A', 'issues'). The plot is interactive, using Plotly for visualization. The function determines the title and axis labels based on the presence of specific columns in the DataFrame.

    Parameters:
    df (pd.DataFrame): A pandas DataFrame containing the time series data. The DataFrame's index should represent the timestamp for each data point.

    Returns:
    None: The function directly displays the plot using fig.show() and does not return any value.
    """
    fig = go.Figure()
    for col in df.columns:
        if col not in ['V','A','issues']:
            fig.add_trace(go.Scatter(x=df.index, y=df[col], mode='lines', name=col)) 
    if "P_agg" in df.columns:
        fig.update_layout(title='Consumption Data',
                  xaxis_title='Timestamp',
                  yaxis_title='Power (W)')
    else:
         fig.update_layout(title='Environmental Data',
                  xaxis_title='Timestamp',
                  yaxis_title='')
    fig.show()
    
    
make_plot(electric_data)
```

### Environmental data plots

```python
house = 'House_03'
month = '2023-08'

# Parse data
# ----------
env_data = pd.read_csv('clean_dataset/'+house+'/Environmental_data/'+month+'.csv',header=0)
env_data['timestamp'] = pd.to_datetime(env_data['timestamp'])
env_data.set_index('timestamp', inplace=True)
en  = env_data

electric_data = pd.read_csv('clean_dataset/'+house+'/Electric_data/'+month+'.csv',header=0)
electric_data['timestamp'] = pd.to_datetime(electric_data['timestamp'])
electric_data.set_index('timestamp', inplace=True)
ac = electric_data[['ac_1']]

en = env_data.loc['2023-08-24 00:00:00':'2023-08-30 08:00:00']
ac = ac.loc['2023-08-24 00:00:00':'2023-08-30 08:00:00']


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(21, 10))
line_styles1 = ['-', '-', '--', '--']
line_widths1 = [4.0, 3.0, 3.0, 4.0]
line_colors1 = ['#EBD3A9', '#99AFB8', '#3282A6', '#EDB699']
legend_labels1 = ['internal temp (C)', 'internal  humidity (%)', 'external temp (C)', 'external humidity (%)']

for i, column in enumerate(en.columns):
     ax1.plot(en.index, en[column], label=legend_labels1[i], linestyle=line_styles1[i], linewidth=line_widths1[i], color=line_colors1[i])

ax1.grid(True, linestyle='--', alpha=0.4)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.set_title('Environmental data - AC power', fontsize=27)
ax1.xaxis.set_major_locator(MaxNLocator(nbins=5))
ax1.tick_params(axis='both', which='major', labelsize=18)
legend1 = ax1.legend(frameon=False, facecolor='gray', framealpha=1, fontsize=12, loc ='upper right')
ax1.legend(loc='center left', bbox_to_anchor=(0.95, 0.83),prop={'size': 20},frameon=False)

# Third subplot (bottom-left)
# Add your third plot (e.g., ax3.plot(x3, y3, label='Data3', color='blue')) here
ax2.plot(ac.index, ac['ac_1'], label='A/C', color='#FF8685', linewidth=3)
ax2.fill_between(ac.index, ac['ac_1'], color='#FF8685', alpha=0.5)  # Fill the area below the line
ax2.set_ylabel('Power (Watts)',fontsize=20)
ax2.grid(linestyle='--', alpha=0.4)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.xaxis.set_major_locator(MaxNLocator(nbins=5))
ax2.tick_params(axis='both', which='major', labelsize=18)
ax2.legend(loc='upper right', bbox_to_anchor=(1.16, 1),prop={'size': 20},frameon=False)

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```
